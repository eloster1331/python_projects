```python
jumbled_list_in_need_of_sorting = input("Behold, I desire the jumbled, comma seperated lists! ")
split_list = jumbled_list_in_need_of_sorting.split(",")
clean_stripped_list = [item.strip() for item in split_list]
clean_stripped_list.sort()
output_list = ", ".join(clean_stripped_list)
print(output_list)




Silly_question = input("Wanna have every other?" )
if Silly_question == "yes":
  print(clean_stripped_list[::2])
elif Silly_question == "no":
  print("... ok then?")
```
