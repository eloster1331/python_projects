```python
#FOR LOOPS
numbers = [1, 2, 3, 4, 5, 6]
for item in numbers:
  print(item)
print("BREAK")
i = 0
while i < len(numbers):
  print(numbers[i])
  i = i + 1
```

```python
# RANGE () FUNCTION!

numbers = range(5)
print(numbers)

for number in numbers:
  print(number)

print("Break!")
numbers = range(5, 10)

for number in numbers:
  print(number)

print("Break!")
numbers = range(5, 10, 2)

for number in numbers:
  print(number)

print("Break!")

for number in range(5):
  print(number)
```
