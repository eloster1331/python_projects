```python
def remove_vowels(input_string):
    vowels = "aeiouAEIOU"
    include_y_input = input("Do you want to include 'y' and 'Y' as vowels? (yes/no): ")
    should_include_y = include_y_input.lower() == 'yes'

    if should_include_y:
        vowels += "yY"
    result = ""
    for char in input_string:
        if char not in vowels:
            result += char
    return result

my_result = remove_vowels("cOCEaZyFcBuPqdGxhUeuFbtWVXQjgsZdFUyoaxBSVmeTGRjdnarQEpozeuBWQgTWjANMFGMqoQmJggqMNytOhNwkoOoWzqvVhkSM")

print(my_result)
```
